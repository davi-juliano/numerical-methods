```python
import numpy
import math

def eulerEDO(h,y0,t0,x):
    yn=0.
    yn=y0
    tn=t0
    if(x!=0):
        while(tn<=x):
            yn = yn + h*(0.04*yn)
            tn+=h
    return yn

def rungeKutta2(h,y0,t0,x):
        yn=0.
        yn=y0
        tn=t0
        k=0.
        if(x!=0):
            while(tn<=x):
                s1=0.04*yn
                s2=0.04*(yn+h*s1)
                yn = yn+(h/2)*(s1+s2)
                tn+=h
        return yn

def rungeKutta4(h,y0,t0,x):
        yn=0.
        yn=y0
        tn=t0
        s1=0.
        s2=0.
        s3=0.
        s4=0.
        if(x!=0):
            while(tn<=x):
                s1=(0.04*yn)
                s2=(0.04*(yn+(h/2*s1))) 
                s3=(0.04*(yn+(h/2*s2))) 
                s4=(0.04*(yn+(h*s3))) 
                yn = yn + (h/6)*(s1+2*s2+2*s3+s4)
                tn+=h
        return yn



'''Primeiro Valor de h'''
t0=0
y0=1000
h=0.5
x=1
print("O resultado(letra a, primeiro h) √©",eulerEDO(h,y0,t0,x))  #Letra a
print("O resultado(letra b, primeiro h) √©",rungeKutta2(h,y0,t0,x))  #Letra b
print("O resultado(letra c, primeiro h) √©",rungeKutta4(h,y0,t0,x))  #Letra c


'''Segundo Valor de h'''
t0=0
y0=1000
h=0.25
x=1
print("O resultado(letra a, segundo h) √©",eulerEDO(h,y0,t0,x)) #Letra a
print("O resultado(letra b, segundo h) √©",rungeKutta2(h,y0,t0,x))  #Letra b
print("O resultado(letra c, segundo h) √©",rungeKutta4(h,y0,t0,x))  #Letra c


'''Terceiro Valor de h'''
t0=0
y0=1000
h=0.1
x=1
print("O resultado(letra a, terceiro h) √©",eulerEDO(h,y0,t0,x))  #Letra a
print("O resultado(letra b, terceiro h) √©",rungeKutta2(h,y0,t0,x))  #Letra b
print("O resultado(letra c, terceiro h) √©",rungeKutta4(h,y0,t0,x))  #Letra c
```

```python
def dy_dx(y, x):
    return (2000 - y)/(200 - x)

x = 5
t = np.linspace(0, x, 10000)
y0 = 0
yn = odeint(dy_dx, y0, t)

lista = yn.tolist()
print(lista[9999])

plt.plot(t,yn)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.savefig('grafico7.png', format='png')
plt.show()
```

```python
import numpy
import math

def eulerEDO(h,y0,t0,x):
    yn=0.
    yn=y0
    tn=t0
    if(x!=0):
        while(tn<=x):
            yn = yn + h*(2000-2*yn)/(200-tn)
            tn+=h
    return yn

def solexataEDO(x):
    yn=(10*x)+(1/40)*(x**2)
    return yn

def rungeKutta4(h,y0,t0,x):
        yn=0.
        yn=y0
        tn=t0
        s1=0.
        s2=0.
        s3=0.
        s4=0.
        if(x!=0):
            while(tn<=x):
                s1=(2000-yn)/(200-tn)
                s2=(2000-yn+s1*h/2)/(200-tn+h/2) 
                s3=(2000-yn+s2*h/2)/(200-tn+h/2) 
                s4=(2000-yn+s3*h)/(200-tn+h) 
                yn = yn + (h/6)*(s1+2*s2+2*s3+s4)
                tn+=h
        return yn



t0=0
y0=0
h=0.5
x=5
print("Solu√ß√£o exata: ", solexataEDO(x))
print("(a) Metodo de Euler: ", eulerEDO(h,y0,t0,x))
print("(b) Runge-Kutta", rungeKutta4(h,y0,t0,x))
```
